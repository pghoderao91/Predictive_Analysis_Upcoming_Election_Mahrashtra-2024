**4.1] Univariate Analysis**

```python
## Winning votes

data['Winning Vote'] = data['Winning Vote'].astype(int)

def remove_commas(value):
    if isinstance(value, str) and ',' in value:
        return value.replace(',', '')
    return value

# Apply the function to the 'Total Electors' column
data['Total Votes'] = data['Total Votes'].apply(remove_commas)

# Convert the 'Total Electors' column to numeric, handling NaNs appropriately
data['Total Votes'] = pd.to_numeric(data['Total Votes'], errors='coerce')

data['Margin'] = data['Margin'].replace('[^0-9]', '', regex=True)
data['Margin'] = pd.to_numeric(data['Margin'], errors='coerce')

data['Total Electors'] = data['Total Electors'].astype(str)
data['Total Electors'] = data['Total Electors'].str.replace(',', '')
data['Total Electors'] = pd.to_numeric(data['Total Electors'], errors='coerce')

data['Year_datetime'] = pd.to_datetime(data['Year'], format='%Y-%d-%m', errors='coerce')
data.drop(columns=['Year'], inplace=True)
```

**7] Model Building and Evaluation**

**1]Defining The independent and Dependent Variable**

```python
data2
```

```python
x= data2.drop(columns=['Winning Vote'])
y= data2['Winning Vote']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x, y,test_size=0.15,random_state=42)
```

```python
print('x_train.shape:',x_train.shape)
print('x_test.shape:',x_test.shape)
print('y_train.shape:',y_train.shape)
print('y_test.shape:',y_test.shape)
```

**Linear Regression**

```python
# Initialize and train model
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
lr = LinearRegression()
lr.fit(x_train,y_train)

# Predict and Evaluate the model

y_pred = lr.predict(x_test)

mse=mean_squared_error(y_test,y_pred)
print('MSE:',mse)

R2_score = r2_score(y_test,y_pred)
print('R_2Score:',R2_score)

adjusted_r2= 1-(1-R2_score)*(216-1)/(216-8-1)
print('Adjusted_r2_score:',adjusted_r2)
```

**Support Vector Regressor**

```python
from sklearn.svm import SVR
svr = SVR()
svr.fit(x_train,y_train)

# Predict and Evaluate the model

y_pred1 = svr.predict(x_test)

mse=mean_squared_error(y_test,y_pred1)
print('MSE:',mse)

R2_score = r2_score(y_test,y_pred1)
print('R_2Score:',R2_score)

adjusted_r2= 1-(1-R2_score)*(216-1)/(216-8-1)
print('Adjusted_r2_score:',adjusted_r2)
```

**Hyper-Parameter Tunning for SVR**

```python
#define the parameters
param_grid = {
    'C': [0.1, 1, 10, 100],            # Regularization parameter
    'epsilon': [0.01, 0.1, 1],         # Epsilon in the epsilon-SVR model
    'kernel': ['linear', 'poly', 'rbf', 'sigmoid'],  # Kernel type to be used in the algorithm
    'gamma': ['scale', 'auto']         # Kernel coefficient
}
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVR
svr = SVR()

grid_search = GridSearchCV(estimator=svr, param_grid=param_grid, cv=5, scoring='r2', verbose=2, n_jobs=-1)
grid_search.fit(x_train, y_train)
```

```python
print("Best parameters:", grid_search.best_params_)
```

```python
model1 = SVR(C=100,epsilon=0.01,gamma='scale',kernel='linear')
model1.fit(x_train,y_train)

y_pred_svr = model1.predict(x_test)
y_pred_svr

mae=mean_absolute_error(y_test,y_pred_svr)
print('MAE:',mae)

r_score = r2_score(y_test,y_pred_svr)
print('Hyper_Parameter_R_2Score:',r_score)

adjusted_r2= 1-(1-r_score)*(216-1)/(216-8-1)
print('Adjusted_r2_score:',adjusted_r2)
```

**Decision Tree Regressor**

```python
from sklearn.tree import DecisionTreeRegressor
dt = DecisionTreeRegressor()
dt.fit(x_train,y_train)

y_pred2 = dt.predict(x_test)

Mse=mean_squared_error(y_test,y_pred2)
print('MSE:',Mse)

R_2Score = r2_score(y_test,y_pred2)
print('R_2Score:',R_2Score)

Adjusted_r2_score = 1-(1-R_2Score)*(216-1)/(216-8-1)
print('Adjusted_r2_score:',Adjusted_r2_score)
```

```python
from keras.models import Sequential
from keras.layers import Dense
import numpy as np

## initialize the ann model

ann_model = Sequential()


## add layer (example)

ann_model.add(Dense(128, input_dim=x_train.shape[1],activation='tanh'))
ann_model.add(Dense(64,activation='relu'))
ann_model.add(Dense(1,activation='linear'))

ann_model.add(Dense(128,activation='relu'))
ann_model.add(Dense(64,activation='relu'))
ann_model.add(Dense(1,activation='linear'))

ann_model.add(Dense(128,activation='relu'))
ann_model.add(Dense(64,activation='relu'))
ann_model.add(Dense(1,activation='linear'))



ann_model.compile(optimizer='adam',loss='mean_squared_error',metrics=['mae'])
ann_model.summary()

## train the model
ann_model.fit(x_train,y_train,epochs=100,batch_size=32,verbose=0)

y_pred_ann = ann_model.predict(x_test)

# Calculate metrics
mse_ann = mean_squared_error(y_test, y_pred_ann)
rmse_ann = np.sqrt(mse_ann)
r2_ann = r2_score(y_test, y_pred_ann)

print("Mean Squared Error (MSE):", mse_ann)
print("Root Mean Squared Error (RMSE):", rmse_ann)
print("R-squared (R2) Score:", r2_ann)
```
